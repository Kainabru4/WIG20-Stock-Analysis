```python
#importing packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import plotly.graph_objects as go
import plotly.io as pio
```

#### Daily Data - 5y

```python
daily = pd.read_csv('wig20_5d.csv') #reading data 
daily['Data'] = pd.to_datetime(daily['Data']) #changing date format
daily.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume'] #renaming columns
daily.head()
```

```python
daily.iloc[:,1:].describe().round(2) #calculating basic statistics
```

```python
#creating a plot of 'open' price against time

daily.plot(x='Date', y='Open', color='navy', linewidth=2)

ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

plt.xticks(rotation=90)
plt.xlabel('Data')
plt.ylabel('Cena Otwarcia')
plt.show()
```

```python
pio.renderers.default = 'browser' #ensuring following the plot opens in web
```

```python
#creating candlestick plot

fig = go.Figure(data=[go.Candlestick(x = daily['Date'],
                open = daily['Open'],
                high = daily['High'],
                low = daily['Low'],
                close = daily['Close'])])

fig.show()
```
