```python
#Imports packages
import pandas as pd
import numpy as np
```

```python
#Reads data
indicators = pd.read_csv('../data/processed/technical_indicators.csv')
#Slices data
indicators = indicators.iloc[:, 1:]
#Names columns
indicators.columns = [
    'Date', 
    'Open', 
    'High', 
    'Low', 
    'Close', 
    'Volume', 
    'RSI', 
    'Stochastic Oscillator', 
    'SMAs',
    'SMAm', 
    'MACD', 
    'A/D'
    ]
indicators.head()
```

```python
#Ceates a column with additional metric 'Target'
indicators['Target'] = indicators['Close']-indicators.Open
indicators['Target'] = indicators['Target'].shift(-1)

#Ceates a column with binary metric 'TargetClass'
indicators['TargetClass'] = [1 if indicators.Target[i]>0 else 0 for i in range(len(indicators))]
#Ceates a column with additional metric 'TargetNextClose'
indicators['TargetNextClose'] = indicators['Close'].shift(-1)
#Drops missing data
indicators.dropna(inplace=True)

indicators.head()
```

```python
#Reads data
sentiment = pd.read_csv('../data/processed/sentiment.csv')
#Slices data
sentiment = sentiment.iloc[:, 1:]
#Reames column
sentiment.rename(columns={'date': 'Date'}, inplace=True)
sentiment.head()
```

```python
#Sets a timespan of 1y
indicators_1y = indicators[(indicators['Date'] >= '2023-06-01') & (indicators['Date'] <= '2024-08-30')]
#Changes index to date
indicators_1y.set_index('Date', inplace=True)
indicators_1y.head()
```

```python
#Saves dataset
indicators_1y.to_csv('../data/processed/dataset_1y.csv')
```

```python
#Sets a timespan of 5y
indicators_5y = indicators[(indicators['Date'] >= '2019-06-01') & (indicators['Date'] <= '2024-08-30')]
#Changes index to date
indicators_5y.set_index('Date', inplace=True)
indicators_5y.head()
```

```python
#Saves dataset
indicators_5y.to_csv('../data/processed/dataset_5y.csv')
```

```python
#Merges datasets on date if in both columns
indicators_merged = pd.merge(indicators, sentiment, on='Date', how='inner')
#Changes index to date
indicators_merged.set_index('Date', inplace=True)
#Drops column text
indicators_merged = indicators_merged.drop(columns=['text'])

#Renames columns
indicators_merged.rename(columns=
    {
    'sentiment_score_2_fold': 'Binary', 
    'sentiment_score_3_fold': 'Neutral',
    'sentiment_score_SO_PMI': 'SO_PMI'
    }, 
    inplace=True)
columns = [
    'Open', 
    'High', 
    'Low', 
    'Close', 
    'Volume', 
    'RSI', 
    'Stochastic Oscillator', 
    'SMAs',
    'SMAm', 
    'MACD', 
    'A/D',
    'Binary', 
    'Neutral', 
    'SO_PMI',
    'Target', 
    'TargetClass', 
    'TargetNextClose'
]
indicators_merged = indicators_merged[columns]

indicators_merged.head()
```

```python
#Creates 3 different datasets with a single sentiment metric each
indicators_binary = indicators_merged.drop(columns=['Neutral', 'SO_PMI'])
indicators_neutral = indicators_merged.drop(columns=['Binary', 'SO_PMI'])
indicators_so_pmi = indicators_merged.drop(columns=['Binary', 'Neutral'])
```

```python
#Saves datasets
indicators_binary.to_csv('../data/processed/dataset_binary.csv')
indicators_neutral.to_csv('../data/processed/dataset_neutral.csv')
indicators_so_pmi.to_csv('../data/processed/dataset_so_pmi.csv')
```
